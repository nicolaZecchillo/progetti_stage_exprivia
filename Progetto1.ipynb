```python
import pandas as pd
```

```python
df = pd.read_csv('../dati/POSAS_2024_it_Province.csv', sep=';')
df.head()
df.count
```

```python
df_iso = pd.read_csv('../dati/ISO_Regioni_Province.csv')
df_iso.head()
df_iso.count
```

```python
df_iso.head()
```

df_conISO = df_conISO.drop('Provincia', axis=1)

```python
df_conISO = df_iso.merge(df, left_on='den_geografica', right_on='Provincia' , how='right')
df_conISO = df_conISO.drop('Codice provincia', axis=1)
df_conISO.loc[df_conISO['den_geografica'] == 'Napoli', 'ISO'] = df_conISO.loc[df_conISO['den_geografica'] == 'Napoli', 'ISO'].fillna('IT-NA')
df_conISO.head()
```
