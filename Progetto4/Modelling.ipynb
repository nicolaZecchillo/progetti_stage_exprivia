```python
df2023 = pd.read_csv('input/Popolazione_residente_2023.csv')
df2022 = pd.read_csv('input/Popolazione_residente_2022.csv')
df2021 = pd.read_csv('input/Popolazione_residente_2021.csv')
df2020 = pd.read_csv('input/Popolazione_residente_2020.csv')
df2019 = pd.read_csv('input/Popolazione_residente_2019.csv')
```

1. Creo due liste contenenti i dataFrames e l'elenco degli anni, una vuota che andrà a memorizzare i df modificati al termine del ciclo for 
2. Nel ciclo creo la colonna anno con il valore del ciclo corrente e appenendo il df alla lista df_trasformati.
3. Concateno tutti i df nella lista, rimuovo tutte le colonne duplicate a preservando la prima trovata [:, ~], sostituisco il valore '100 e oltre' con il valore 100 ed escludo quelle in cui il valore 'Età' è 'Totale'

```python
dataframes = [df2019, df2020, df2021, df2022, df2023, df2024]
anni = [2019, 2020, 2021, 2022, 2023, 2024]

df_trasformati = []

for df, anno in zip(dataframes, anni):
    df = df.copy()
    df['Anno'] = anno

    df_trasformati.append(df)

df_2020_2024 = pd.concat(df_trasformati, axis=0)
df_2020_2024 = df_2020_2024.loc[:, ~df_2020_2024.columns.duplicated()]
df_2020_2024['Età'] = df_2020_2024['Età'].replace('100 e oltre', 100)
df_2020_2024 = df_2020_2024[df_2020_2024['Età'] != 'Totale']
df_2020_2024.head()
```

```python
df_2020_2024.to_csv('outputModelling/Popolazione_residente_2019-2024.csv', index=False)
```

```python
df_2002_2019 = pd.read_csv('input/Ricostruzione della popolazione 2002-2019.csv')
df_2002_2019.head()
```

1. Divido il df in due df sulla base dei diversi valori della colonna 'Sesso'
2. Attraverso il metodo .melt() rendo le colonne con gli anni le righe di una nuova colonna 'Anno' sui criteri 'Età' e 'Sesso'
3. Riunisco i due df e ricalcolo la colonna 'Totale' come somma
4. Aggiungo la colonna 'Anno' con i valori del melting
5. Seleziono e ordino le colonne necessarie, sostituisco il valore '100 e oltre' con il valore 100 ed escludo quelle in cui il valore 'Età' è 'Totale'

```python
df_maschi = df_2002_2019[df_2002_2019['Sesso'] == 'Maschi']
df_femmine = df_2002_2019[df_2002_2019['Sesso'] == 'Femmine']

df_maschi = df_maschi.melt(id_vars=['Età', 'Sesso'], var_name='Anno', value_name='Totale maschi')
df_femmine = df_femmine.melt(id_vars=['Età', 'Sesso'], var_name='Anno', value_name='Totale femmine')

df_2002_2019_mod = pd.merge(df_maschi, df_femmine, on=['Età', 'Anno'])
df_2002_2019_mod['Totale'] = df_2002_2019_mod['Totale maschi'] + df_2002_2019_mod['Totale femmine']
df_2002_2019_mod['Anno'] = df_2002_2019_mod['Anno'].astype(int)
df_2002_2019_mod = df_2002_2019_mod[['Età', 'Totale maschi', 'Totale femmine', 'Totale', 'Anno']]
df_2002_2019_mod['Età'] = df_2002_2019_mod['Età'].replace('100 e oltre', 100)
df_2002_2019_mod = df_2002_2019_mod[df_2002_2019_mod['Età'] != 'Totale']

df_2002_2019_mod.head()
```
