```python
import pandas as pd
```

```python

df2024 = pd.read_csv('input/Popolazione_residente_2024.csv')
df2024.head()
```

```python
df2023 = pd.read_csv('input/Popolazione_residente_2023.csv')
df2022 = pd.read_csv('input/Popolazione_residente_2022.csv')
df2021 = pd.read_csv('input/Popolazione_residente_2021.csv')
df2020 = pd.read_csv('input/Popolazione_residente_2020.csv')
```

1. Creo due liste contenenti i dataFrames e l'elenco degli anni, una vuota che andrà a memorizzare i df modificati al termine del ciclo for 
2. Nel ciclo divido ogni df in tre diversi, ognuno a seconda di maschi, femmine o totale e creo la colonna 'Sesso' per contenere questo valore, rinomino le colonne 'Totale..' con l'anno preso dalla lista creata in precedenza e ri-unisco in  df in uno unico per ogni anno da aggiungere alla lista df_trasformati.
3. Concateno tutti i df nella lista, rimuovo tutte le colonne duplicate a preservando la prima trovata [:, ~], creo un ordine gerarchico nella colonna 'Sesso' con il metodo Categotical()
4. Ordino il df per 'Età' e 'Sesso' secondo la gerarchia appena creata

```python
dataframes = [df2020, df2021, df2022, df2023, df2024]
anni = ['2020', '2021', '2022', '2023', '2024']

df_trasformati = []

for df, anno in zip(dataframes, anni):
    df_male = df[['Età', 'Totale maschi']].copy()
    df_female = df[['Età', 'Totale femmine']].copy()
    df_total = df[['Età', 'Totale']].copy()

    df_male['Sesso'] = 'Maschi'
    df_female['Sesso'] = 'Femmine'
    df_total['Sesso'] = 'Totale'

    df_male.rename(columns={'Totale maschi': anno}, inplace=True)
    df_female.rename(columns={'Totale femmine': anno}, inplace=True)
    df_total.rename(columns={'Totale': anno}, inplace=True)

    df_modificato = pd.concat([df_male, df_female, df_total])

    df_trasformati.append(df_modificato)

df_2020_2024 = pd.concat(df_trasformati, axis=1)
df_2020_2024 = df_2020_2024.loc[:, ~df_2020_2024.columns.duplicated()]
df_2020_2024 = df_2020_2024[['Età', 'Sesso'] + [anno for anno in anni]]
df_2020_2024['Sesso'] = pd.Categorical(df_2020_2024['Sesso'], categories=['Maschi', 'Femmine', 'Totale'], ordered=True)
df_2020_2024.sort_values(by=['Età', 'Sesso'], inplace=True)
df_2020_2024.head()
```

```python
df_2020_2024.to_csv('outputModelling/Popolazione_residente_2020-2024.csv', index=False)
```

```python
df_2002_2019 = pd.read_csv('input/Ricostruzione della popolazione 2002-2019.csv')
df_2002_2019.head()
```

```python
df_2002_2024 = df_2002_2019.merge(df_2020_2024, left_on=['Età', 'Sesso'], right_on=['Età', 'Sesso'] ,how='inner')
print(df_2002_2024)
```

```python
df_2002_2024.to_csv('outputModelling/Popolazione_residente_2002-2024.csv', index=False)
df_2002_2024 = df_2002_2024[~df_2002_2024['Età'].isin(['100 e oltre', 'Totale'])]
df_2002_2024['Età'] = df_2002_2024['Età'].astype(int)
df_2030 = pd.read_csv('previsioni/Previsione_popolazione_2030.csv')
df_2030[['2025','2026','2027','2028','2029','2030']] = df_2030[['2025','2026','2027','2028','2029','2030']].astype(int)
df_2002_2030 = df_2002_2024.merge(df_2030, left_on=['Età', 'Sesso'], right_on=['Età', 'Sesso'] ,how='inner')
df_2002_2030.to_csv('outputModelling/Popolazione_residente_2002-2030.csv', index=False)
```
