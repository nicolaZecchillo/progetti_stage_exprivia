```python
import pandas as pd
```

```python
ripartizioni = pd.read_csv('output/shp_to_csv/ripartizioni.csv')
ripartizioni['cod_geografico'] = 'IT-' + ripartizioni['COD_RIP'].astype(str)
ripartizioni.rename(columns={'DEN_RIP': 'den_geografica'}, inplace=True)
ripartizioni['tipo_geografico'] = 'Ripartizione'
ripartizioni['parent_id'] = None
num_ripartizioni = len(ripartizioni)
ripartizioni.head()
```

```python
ISO_regioni = pd.read_csv('../input/ISO_Regioni.csv')
ISO_regioni.head()
```

```python
regioni = pd.read_csv('output/shp_to_csv/regioni.csv')
regioni = regioni.merge(ISO_regioni, left_on='DEN_REG', right_on='DEN_REG', how='inner')
regioni.rename(columns={'COD_RIP': 'parent_id', 'DEN_REG': 'den_geografica'}, inplace=True)
regioni['tipo_geografico'] = 'Regione'
num_regioni = len(regioni)
regioni.head()

```

```python
province = pd.read_csv('output/shp_to_csv/province.csv')
province['cod_geografico'] = 'IT-' + province['SIGLA']
province.rename(columns={'COD_REG': 'parent_id', 'DEN_UTS': 'den_geografica'}, inplace=True)
province.loc[province['den_geografica'] == 'Napoli', 'cod_geografico'] = province.loc[province['den_geografica'] == 'Napoli', 'cod_geografico'].fillna('IT-NA')
province['tipo_geografico'] = 'Provincia'
province['parent_id'] = province['parent_id'] + num_ripartizioni
province.head()
```

```python
comuni = pd.read_csv('output/shp_to_csv/comuni.csv', na_values='', keep_default_na=False)
comuni.rename(columns={'COD_PROV': 'parent_id', 'COMUNE': 'den_geografica', 'PRO_COM':'cod_geografico'}, inplace=True)
comuni['tipo_geografico'] = 'Comune'
comuni['parent_id'] = comuni['parent_id'] + num_regioni
comuni.head()
```

```python
comuni.loc[comuni['den_geografica'] == 'None']
```

```python
merge = pd.concat([ripartizioni, regioni, province, comuni], ignore_index=True)
merge['id'] = range(1, len(merge) + 1)
merge = merge[['id','parent_id','tipo_geografico','cod_geografico','den_geografica']]
print(merge)
```

```python
merge.to_csv('output/dim_geografia.csv', index=False)
```
